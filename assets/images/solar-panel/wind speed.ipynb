```python
import pandas as pd
import scipy.stats as stats
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('1732907.csv', index_col=['DATE'], parse_dates=['DATE'])
```

```python
wsfg = df['WSFG'].dropna()
wsfg.shape
```

```python
wsfg.hist(bins=50)
```

```python
q = 0.9999
wsfg.quantile(q)
```

```python
b, loc, scale = stats.lognorm.fit(wsfg)
x = np.linspace(0.0, 100.0, 100)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
ax.plot(x, stats.lognorm.pdf(x, b, loc, scale))
ax.hist(wsfg, bins=50, density=True)
ax.set_ylabel('Wind speed (mph)')
ax.set_ylabel('Probability')
ax.set_title('Chicago daily peak gust wind speed PDF')
ax.grid(True)
```
